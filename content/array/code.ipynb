# [开发商购买土地](https://kamacoder.com/problempage.php?pid=1044)

## 代码

```python
import sys

def main():
    input = sys.stdin.read()
    data = input.split()

    # 获取行列数
    index = 0
    n = int(data[0])
    m = int(data[1])
    index += 2

    # 获取数据vec，顺便在遍历的时候求下total和p(每行的总和)
    vec = []
    row = []
    total = 0
    p = [0] * n # 存储每行的总和
    for i in range(n):
        # 统计每行的数据，append到row中
        row = [] # ! 注意位置别放错
        for _ in range(m):
            num = int(data[index])
            row.append(num)
            p[i] += num
            total += num
            index += 1
        # append到二维数组vec中
        vec.append(row)

    # 行划分
    sum_a = 0
    sum_b = total
    i = 0
    while sum_a < sum_b and i < n:
        sum_a += p[i]
        sum_b -= p[i]
        i += 1
    res_1 = sum_a - sum_b
    res_2 = -1 * res_1 + 2 * p[i-1] # ！注意下标不要越界

    # 列划分
    sum_a = 0
    sum_b = total
    j = 0
    while sum_a < sum_b and j < m:
        # 计算第j列的总和
        total_j = 0
        for i in range(n):
            total_j += vec[i][j]
        sum_a += total_j
        sum_b -= total_j
        j += 1
    res_3 = sum_a - sum_b
    res_4 = -1 * res_3 + 2 * total_j
    result = min(res_1,res_2, res_3, res_4)

    # 输出
    print (result)

if __name__ == "__main__":
    main()
```

# [区间和](https://kamacoder.com/problempage.php?pid=1070)

```python
import sys

input = sys.stdin.read() # input是字符串

def main():
    # 转成列表
    data = input.split()

    # 给出整数数组vec
    index = 0
    n = int(data[0]) # 整数数组的长度 ！注意转成int
    index += 1

    vec = [0] * n # 初始化vec
    for i in range(n):
        vec[i] = int(data[index])
        index += 1 # 跳出循环后的index = 进入循环前的index值 + 循环次数 

    # 记录累加和到p[i]中
    p = [0] * n
    p[0] = vec[0]
    for i in range(n-1):
        p[i+1] = p[i] + vec[i+1]

    # 计算给定区间的和
    results = []
    while index < len(data):
        a = int(data[index])
        b = int(data[index + 1])
        index += 2
        if a == 0:
            sum_value = p[b]
        else:
            sum_value = p[b] - p[a-1] # ！注意下标越界情况
        results.append(sum_value) # ！注意append的用法，易错写为results = results.append(sum_value)

    # 标准输出
    for result in results:
        print(result)

if __name__ == "__main__":
    main()
```
